## simple linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
########### toy data set used in lecture

#X = np.array([[0],[1],[2]])  # x_i values
#y = np.array([1,0,2])  # y_i values

########## simulation  #################
n = 50

X = np.random.uniform(0,10,n);
X = np.sort(X);

y = 3 + 2 * X 

noise = 3 * np.random.randn(n)  # Gaussian noise
# noise = np.arange(0, 5, .05) * np.random.randn(n)   # Gaussian noise with nonconstant variance
# noise = np.random.uniform(-5,5, n)  # uniformly distributed noise
# noise = np.random.exponential(3,n) - 3  # exponentially distributed noise

y = y + noise

plt.scatter(X,y)
```

```python
X = X.reshape((-1, 1))  # convert to column form

reg = LinearRegression().fit(X, y)  # with intercept
```

```python
beta0 = reg.intercept_    # beta_0
beta0
```

```python
beta1 = reg.coef_         # beta_1
beta1
```

```python
Rsq = reg.score(X,y)      # coefficient of determination
Rsq
```

```python
yhat = reg.predict(X)     # fitted values
```

```python
res = y-yhat
```

```python
import pylab 
import scipy.stats as stats

stats.probplot(res, dist="norm", plot=pylab)
pylab.show()
```

```python
# residual plot (against fitted values')
ax = plt.scatter(x=yhat, y=res)
```

```python
reg.predict(np.array([[15]])) # prediction at new location 
```

```python
X1 = np.append(X, np.array([[15]]),axis=0)
y1 = np.append(y,np.array([10]),axis=0)
plt.scatter(X1,y1)
```

```python
reg = LinearRegression().fit(X1, y1)  
reg.coef_   
```
