```python
sum(boston['rm'] > 7)
```

```python
sum(boston['rm'] > 8)
```

```python
above_8 = boston[boston['rm'] > 8]
above_8
```

```python
below_8 = boston[boston['rm'] <= 8]
```

```python
above_8.describe().compare(below_8.describe())
```

Judging by the comparison of the dataframes of above 8 rooms and below 8 rooms, there is a significantly lower crime rate with more than 8 rooms. There is slightly lower property tax rate per $10,000 with more than 8 rooms as well. The lstat is much lower for above_8 as well. There is a much higher MEDV value as well for above_8. This data makes sense when you think about houses with 8 rooms and above will be in nicer suburbs, which will have lower crime and less lower status population, as well as a higher median value of owner-occupied homes.
